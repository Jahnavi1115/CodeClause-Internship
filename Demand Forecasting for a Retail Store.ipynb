```python
import pandas as pd
import matplotlib.pyplot as plt

# Example data setup
# Replace the following line with your actual DataFrame loading logic
train = pd.DataFrame({
    'store': [1, 1, 2, 2],
    'item': [17, 18, 17, 18],
    'date': ['2023-07-01', '2023-07-02', '2023-07-01', '2023-07-02'],
    'sales': [100, 150, 200, 250]
})

forecast = pd.DataFrame({
    'store': [1, 1, 2, 2],
    'item': [17, 18, 17, 18],
    'date': ['2023-07-01', '2023-07-02', '2023-07-01', '2023-07-02'],
    'sales': [110, 160, 210, 260]
})

# Convert 'date' column to datetime if it's not already
train['date'] = pd.to_datetime(train['date'])
forecast['date'] = pd.to_datetime(forecast['date'])

# Check column names to verify the 'item' column exists
print(train.columns)

# Check the first few rows of the DataFrame
print(train.head())

# Plotting the sales and forecast for store 1 item 17
train_filtered = train[(train.store == 1) & (train.item == 17)].set_index("date")
forecast_filtered = forecast[(forecast.store == 1) & (forecast.item == 17)].set_index("date")

fig, ax = plt.subplots(figsize=(20, 9))

train_filtered['sales'].plot(ax=ax, legend=True, label="Store 1 Item 17 Sales")
forecast_filtered['sales'].plot(ax=ax, legend=True, label="Store 1 Item 17 Forecast")

plt.show()
```
